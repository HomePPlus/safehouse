```python
import os
import shutil

# 원본 디렉터리 경로
source_dirs = [
    '/content/강재_강재손상_원천_01/',
    '/content/강재_강재손상_원천_02/'
]

# JSON 파일이 저장된 디렉터리
json_dir = '/content/콘크리트_강재_라벨링/'

# 결과 이미지 저장 경로
target_dir = '/content/콘크리트_강재_원천_2000/'

# 결과 디렉터리 생성
os.makedirs(target_dir, exist_ok=True)

# JSON 파일 이름 추출 (확장자 제거)
json_files = [os.path.splitext(f)[0] for f in os.listdir(json_dir) if f.endswith('.json')]

# 이미지 복사
for source_dir in source_dirs:
    for file_name in os.listdir(source_dir):
        # 파일 이름이 JSON 파일 이름에 포함되며 이미지 파일인지 확인
        if os.path.splitext(file_name)[0] in json_files and file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
            source_file_path = os.path.join(source_dir, file_name)
            target_file_path = os.path.join(target_dir, file_name)
            shutil.copy2(source_file_path, target_file_path)  # 이미지 복사
            print(f"'{source_file_path}' -> '{target_file_path}'로 복사 완료")

print(f"모든 JSON에 대응하는 이미지 파일이 '{target_dir}'에 저장되었습니다.")
```

```python
import os
import json
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# 경로 설정
image_folder = '/content/콘크리트_강재_원천_2000/'  # TIFF 이미지 파일 폴더
json_folder = '/content/콘크리트_강재_라벨링/'  # JSON 파일 폴더

# Step 1: TIFF 이미지 파일 랜덤으로 5개 선택
image_files = [f for f in os.listdir(image_folder) if f.endswith('.tiff')]
selected_images = random.sample(image_files, min(len(image_files), 5))  # 랜덤으로 최대 5개 선택

# Step 2: 이미지와 JSON 내용을 시각적으로 출력
for image_file in selected_images:
    base_name = os.path.splitext(image_file)[0]  # 확장자를 제외한 파일 이름
    json_file = os.path.join(json_folder, base_name + '.json')
    image_path = os.path.join(image_folder, image_file)

    # TIFF 이미지 출력
    img = Image.open(image_path)
    fig, ax = plt.subplots(1, figsize=(8, 8))
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(f"{image_file}")

    # JSON 파일에서 bounding box 가져오기
    if os.path.exists(json_file):
        with open(json_file, 'r') as f:
            json_data = json.load(f)

        annotations = json_data.get('annotations', [])  # 모든 annotations 사용

        # Bounding box 그리기
        for annotation in annotations:
            bbox = annotation.get('bbox', [])
            attributes = annotation.get('attributes', {})
            class_name = attributes.get('class', 'Unknown')  # attributes 안에 class 가져오기

            if len(bbox) == 4:  # bbox는 [x, y, width, height] 형식이어야 함
                x, y, width, height = bbox
                rect = patches.Rectangle(
                    (x, y), width, height, linewidth=2, edgecolor='red', facecolor='none'
                )
                ax.add_patch(rect)
                ax.text(
                    x, y - 5, class_name,  # class_name 표시
                    color='red', fontsize=10, bbox=dict(facecolor='white', alpha=0.5)
                )
    else:
        print(f"{base_name}.json 파일이 없습니다.")

    plt.show()
```

```python
# 경로 설정
json_folder = '/content/콘크리트_강재_라벨링/'  # JSON 파일이 있는 폴더
output_folder = '/content/steel_json/'  # 필터링된 JSON 저장 폴더

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# JSON 파일 처리
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)
        output_json_path = os.path.join(output_folder, json_file)  # 필터링 결과 저장 경로

        try:
            # JSON 데이터 로드
            with open(input_json_path, 'r') as f:
                coco_data = json.load(f)

            # 필터링 작업: SteelDefect 클래스만 포함
            filtered_annotations = [
                annotation for annotation in coco_data.get("annotations", [])
                if annotation.get("attributes", {}).get("class") == "SteelDefect"
            ]

            # 필터링된 데이터 구성
            filtered_coco_data = {
                "info": coco_data.get("info", {}),
                "licenses": coco_data.get("licenses", []),
                "images": coco_data.get("images", []),  # 모든 이미지 유지
                "categories": coco_data.get("categories", []),
                "annotations": filtered_annotations,  # SteelDefect 클래스만 포함
            }

            # 필터링된 JSON 저장
            with open(output_json_path, 'w') as f:
                json.dump(filtered_coco_data, f, ensure_ascii=False, indent=4)

            print(f"{json_file} 처리 완료! 필터링된 결과가 {output_json_path}에 저장되었습니다.")
        except json.JSONDecodeError as e:
            print(f"JSONDecodeError: {json_file}에서 오류 발생. 메시지: {e}")

print("모든 JSON 파일 처리 완료!")
```

```python
from collections import defaultdict

# 경로 설정
json_folder = '/content/steel_json/'  # JSON 파일이 있는 폴더

# 클래스별 개수를 저장할 딕셔너리
class_counts = defaultdict(int)

# JSON 파일 순회
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)

        # JSON 데이터 로드
        with open(input_json_path, 'r') as f:
            coco_data = json.load(f)

        # annotations에서 class 개수 카운트
        for annotation in coco_data.get("annotations", []):
            class_name = annotation.get("attributes", {}).get("class")
            if class_name:  # 클래스가 존재할 경우만 카운트
                class_counts[class_name] += 1

# 결과 출력
print("존재하는 클래스와 각 클래스의 개수:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}개")

# 총 클래스 종류
print(f"\n총 클래스 종류: {len(class_counts)}개")
```

```python
import os
import json
import random
from PIL import Image

# JSON 파일들이 저장된 디렉토리
label_dirs = [
    "/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/라벨링데이터/steel_json/"
]
# 원본 이미지 디렉토리
source_dirs = [
    "/content/강재_강재손상_원천_2000/"
]
# 데이터셋 저장 경로
dataset_dir = "/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/dataset"

# 데이터셋 저장 디렉토리 설정
train_images_dir = os.path.join(dataset_dir, "images", "train")
val_images_dir = os.path.join(dataset_dir, "images", "val")
train_labels_dir = os.path.join(dataset_dir, "labels", "train")
val_labels_dir = os.path.join(dataset_dir, "labels", "val")

# 디렉토리 생성
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# YOLO 라벨 변환 함수 (class_id와 level을 통합)
def convert_bbox_to_yolo_format_with_combined_class(bbox, class_id, level, image_width, image_height, max_level=3):
    try:
        # class_id와 level을 통합하여 고유 클래스 ID 생성
        combined_class_id = (class_id - 1) * max_level + (level - 1)  # 예: class_id=1, level=2 -> combined_class_id=1
        x_min, y_min, bbox_width, bbox_height = bbox
        x_center = (x_min + bbox_width / 2) / image_width
        y_center = (y_min + bbox_height / 2) / image_height
        width = bbox_width / image_width
        height = bbox_height / image_height
        return f"{combined_class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}"
    except Exception as e:
        print(f"⚠️ 바운딩 박스 변환 오류: {bbox}, 오류: {e}")
        return None

# 데이터셋 구축 함수
def create_yolo_dataset_with_combined_class(label_dirs, source_dirs, train_ratio=0.8):
    json_files = []

    # 모든 JSON 파일과 원본 이미지 파일 매칭
    for label_dir, source_dir in zip(label_dirs, source_dirs):
        for file_name in os.listdir(label_dir):
            if file_name.endswith(".json"):
                json_files.append((label_dir, source_dir, file_name))

    random.shuffle(json_files)  # 데이터 섞기

    train_count = int(len(json_files) * train_ratio)
    train_files = json_files[:train_count]
    val_files = json_files[train_count:]

    train_image_count, val_image_count = 0, 0
    train_label_count, val_label_count = 0, 0

    for label_dir, source_dir, json_file in json_files:
        json_path = os.path.join(label_dir, json_file)
        try:
            with open(json_path, 'r', encoding='utf-8') as f:
                data = json.load(f)

            # 이미지 정보 가져오기
            image_info = data["images"][0]
            image_path = os.path.join(source_dir, image_info["file_name"])
            image_width, image_height = image_info["width"], image_info["height"]

            # 저장 경로 설정
            if (label_dir, source_dir, json_file) in train_files:
                image_save_dir = train_images_dir
                label_save_dir = train_labels_dir
                train_image_count += 1
            else:
                image_save_dir = val_images_dir
                label_save_dir = val_labels_dir
                val_image_count += 1

            # 이미지 저장 (.tiff -> .jpg 확장자 변경)
            dest_image_path = os.path.join(image_save_dir, image_info["file_name"].replace(".tiff", ".jpg"))
            if not os.path.exists(dest_image_path):
                with Image.open(image_path) as img:
                    img.convert("RGB").save(dest_image_path, "JPEG")

            # YOLO 라벨 저장
            label_save_path = os.path.join(label_save_dir, image_info["file_name"].replace(".tiff", ".txt"))
            with open(label_save_path, 'w') as f:
                for annotation in data["annotations"]:
                    bbox = annotation["bbox"]
                    class_id = annotation.get("category_id", 1)  # 기본 class_id
                    level = annotation.get("level", 1)  # 기본 level
                    yolo_label = convert_bbox_to_yolo_format_with_combined_class(
                        bbox, class_id, level, image_width, image_height
                    )
                    if yolo_label:  # 변환 성공한 경우만 저장
                        f.write(f"{yolo_label}\n")
                        if (label_dir, source_dir, json_file) in train_files:
                            train_label_count += 1
                        else:
                            val_label_count += 1
                    else:
                        print(f"⚠️ 라벨 변환 실패: {json_file}, 바운딩 박스: {bbox}")

        except Exception as e:
            print(f"⚠️ 오류 발생: {json_file}, 오류: {e}")

    # 결과 출력
    print(f"Train 데이터셋 - 이미지: {train_image_count}, 라벨: {train_label_count}")
    print(f"Validation 데이터셋 - 이미지: {val_image_count}, 라벨: {val_label_count}")

# 데이터셋 생성 실행
create_yolo_dataset_with_combined_class(label_dirs, source_dirs)
```

```python
import os
import json
from collections import Counter

# JSON 파일들이 저장된 디렉토리
label_dirs = [
    "/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/라벨링데이터/steel_json/"
]

# Level 카운트를 저장할 Counter 객체
level_counts = Counter()

# JSON 파일 순회
for label_dir in label_dirs:
    for file_name in os.listdir(label_dir):
        if file_name.endswith(".json"):
            json_path = os.path.join(label_dir, file_name)
            try:
                with open(json_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    # annotations에서 level 값 추출 및 카운트
                    for annotation in data.get("annotations", []):
                        level = annotation.get("level", 0)
                        level_counts[level] += 1
            except Exception as e:
                print(f"⚠️ 오류 발생: {file_name}, 오류: {e}")

# 결과 출력
for level, count in sorted(level_counts.items()):
    print(f"Level {level}: {count}개")
```

```python
import os

# 원본 이미지 디렉토리
image_dir = "/content/강재_강재손상_원천_2000/"

# 삭제된 파일 경로 저장
deleted_images = []

for json_file in json_deleted_files:
    image_name = os.path.basename(json_file).replace(".json", ".tiff")
    image_path = os.path.join(image_dir, image_name)
    print(f"Checking: {image_path}")
    if os.path.exists(image_path):
        try:
            os.remove(image_path)
            print(f"Deleted: {image_path}")
            deleted_images.append(image_path)
        except Exception as e:
            print(f"Error deleting: {image_path}, Error: {e}")
    else:
        print(f"File not found: {image_path}")


# 결과 출력
print(f"총 {len(deleted_images)}개의 이미지 파일이 삭제되었습니다.")
for image in deleted_images:
    print(image)
```

```python
import os
import json
import random

# JSON 파일들이 저장된 디렉토리
label_dirs = [
    "/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/라벨링데이터/steel_json/"
]

# Level 0과 Level 12 파일명을 저장할 리스트
level_0_files = []
level_12_files = []

# JSON 파일 순회
for label_dir in label_dirs:
    for file_name in os.listdir(label_dir):
        if file_name.endswith(".json"):
            json_path = os.path.join(label_dir, file_name)
            try:
                with open(json_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    # Level 값 확인
                    for annotation in data.get("annotations", []):
                        level = annotation.get("level", 0)
                        if level == 0:
                            level_0_files.append(json_path)
                        elif level == 12:
                            level_12_files.append(json_path)
            except Exception as e:
                print(f"⚠️ 오류 발생: {file_name}, 오류: {e}")

# Level 0과 Level 12 파일 중 각각 하나씩 랜덤으로 선택
random_level_0_file = random.choice(level_0_files) if level_0_files else None
random_level_12_file = random.choice(level_12_files) if level_12_files else None

# JSON 내용 확인
if random_level_0_file:
    print(f"랜덤으로 선택된 Level 0 파일: {random_level_0_file}")
    with open(random_level_0_file, 'r', encoding='utf-8') as f:
        level_0_data = json.load(f)
    print("Level 0 파일 내용 예시:")
    print(json.dumps(level_0_data, indent=4, ensure_ascii=False))

if random_level_12_file:
    print(f"랜덤으로 선택된 Level 12 파일: {random_level_12_file}")
    with open(random_level_12_file, 'r', encoding='utf-8') as f:
        level_12_data = json.load(f)
    print("Level 12 파일 내용 예시:")
    print(json.dumps(level_12_data, indent=4, ensure_ascii=False))
```

```python
from ultralytics import YOLO

# YOLO 모델 초기화
model = YOLO('yolo11m.pt')  # 사전 학습된 YOLOv8n 모델

# 모델 학습 실행
model.train(
    data='/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/dataset/dataset.yaml',  # yaml 파일 경로
    epochs=50,                     # 학습 에포크 수
    batch=16,                      # 배치 크기
    imgsz=640,                     # 이미지 크기
    project='/content/yolo_training',  # 학습 결과 저장 경로
    name='steel_defect_detection3',     # 실험 이름
    optimizer='AdamW',             # 옵티마이저 설정 (AdamW 추천)
    lr0=0.001,                     # 초기 학습률
    cos_lr=True,                   # 코사인 학습률 스케줄러
    augment=True,                  # 데이터 증강 활성화
    mosaic=1.0,                    # 모자이크 증강
    hsv_h=0.015,                   # 색조 조정
    hsv_s=0.7,                     # 채도 조정
    hsv_v=0.4,                     # 밝기 조정
    translate=0.1,                 # 평행 이동
    scale=0.5,                     # 확대/축소
    shear=2.0,                     # 기울이기
    flipud=0.5,                    # 상하 뒤집기
    fliplr=0.5                     # 좌우 뒤집기
)
```

```python
import shutil
import os

# YOLO 학습 결과 경로
source_path = "/content/yolo_training/steel_defect_detection32"

# Google Drive 저장 경로
destination_path = "/content/drive/MyDrive/Homeplus/data/Steel/SteelDamage/steel_defect_detection32"

# 대상 폴더 생성
os.makedirs(destination_path, exist_ok=True)

# YOLO 결과 파일 복사
for item in os.listdir(source_path):
    src_item = os.path.join(source_path, item)
    dest_item = os.path.join(destination_path, item)

    # 파일이나 폴더를 구분하여 복사
    if os.path.isfile(src_item):
        shutil.copy2(src_item, dest_item)
    elif os.path.isdir(src_item):
        shutil.copytree(src_item, dest_item, dirs_exist_ok=True)

print(f"YOLO 학습 결과가 {destination_path}에 저장되었습니다.")
```
