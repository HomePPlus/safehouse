# 이미지와 라벨링 파일 처리

## 1) 라벨링 한 2000개의 json 파일 압축 해제하기
- /content/백태누수_2000 폴더에 모두 저장함

```python
import zipfile
import os

zip_path = file_path + '라벨링데이터/라벨링 2000'
extract_path = '/content/백태누수_라벨_2000/'

# zip_path 내의 모든 파일 목록 가져오기
file_list = os.listdir(zip_path)

# .zip 확장자를 가진 파일만 필터링
zip_files = [f for f in file_list if f.endswith('.zip')]

# 최대 4개의 zip 파일 처리 (문제 조건에 따라 4개로 제한)
for zip_file in zip_files:
    zip_file_path = os.path.join(zip_path, zip_file)

    # zip 파일 열기
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # 압축 해제
        zip_ref.extractall(extract_path)

    print(f"'{zip_file}' 압축 해제 완료")

print("모든 zip 파일 압축 해제 완료")


file_count = len([f for f in os.listdir(extract_path) if os.path.isfile(os.path.join(extract_path, f))])

print(f"'{extract_path}' 폴더 내 파일 개수: {file_count}")
```

## JSON 필터링

```python
import json
# 경로 설정
json_folder = '/content/백태누수_라벨_2000/'  # JSON 파일이 있는 폴더
output_folder = '/content/efflorescence_waterLeakage_json/'  # 필터링된 JSON 저장 폴더

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# JSON 파일 처리
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)
        output_json_path = os.path.join(output_folder, json_file)  # 필터링 결과 저장 경로

        try:
            # JSON 데이터 로드
            with open(input_json_path, 'r') as f:
                coco_data = json.load(f)

            # 필터링 작업: Efflorescence 클래스만 포함
            filtered_annotations = [
                annotation for annotation in coco_data.get("annotations", [])
                if annotation.get("attributes", {}).get("class") == "Efflorescene"
            ]

            # 필터링된 데이터 구성
            filtered_coco_data = {
                "info": coco_data.get("info", {}),
                "licenses": coco_data.get("licenses", []),
                "images": coco_data.get("images", []),  # 모든 이미지 유지
                "categories": coco_data.get("categories", []),
                "annotations": filtered_annotations,  # Efflorescence 클래스만 포함
            }

            # 필터링된 JSON 저장
            with open(output_json_path, 'w') as f:
                json.dump(filtered_coco_data, f, ensure_ascii=False, indent=4)

            print(f"{json_file} 처리 완료! 필터링된 결과가 {output_json_path}에 저장되었습니다.")
        except json.JSONDecodeError as e:
            print(f"JSONDecodeError: {json_file}에서 오류 발생. 메시지: {e}")

print("모든 JSON 파일 처리 완료!")
```

```python
import os

# 라벨 파일 경로
label_dir = "/content/drive/MyDrive/Homeplus/data/Concrete/Efflorescence_WaterLeakage/dataset/labels/train"  # 라벨 디렉토리

# 라벨 파일의 컬럼 개수 확인 함수
def check_label_columns(label_dir):
    label_files = [f for f in os.listdir(label_dir) if f.endswith(".txt")]
    print(f"총 {len(label_files)}개의 라벨 파일이 발견되었습니다.\n")

    for label_file in label_files:
        file_path = os.path.join(label_dir, label_file)
        with open(file_path, 'r') as f:
            lines = f.readlines()

        for idx, line in enumerate(lines):
            columns = line.strip().split()  # 공백 기준으로 컬럼 분리
            column_count = len(columns)
            if column_count != 5:  # 5개가 아닌 경우 경고 출력
                print(f"파일: {label_file}, 라인 {idx + 1}: {column_count}개의 컬럼 -> {columns}")

# 실행
check_label_columns(label_dir)
```
