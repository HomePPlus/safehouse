```python
import os
import json
from tqdm import tqdm  # 진행 상황 표시를 위한 라이브러리

# 경로 설정
json_folder = '/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/dataset/jsons'
output_folder = '/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/laveling'

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# JSON 파일 처리
json_files = [f for f in os.listdir(json_folder) if f.endswith('.json')]
for json_file in tqdm(json_files, desc="JSON 파일 처리"):  # 진행 상황 표시
    try:
        input_json_path = os.path.join(json_folder, json_file)
        output_json_path = os.path.join(output_folder, json_file)

        with open(input_json_path, 'r') as f:
            coco_data = json.load(f)

        filtered_annotations = [
            annotation for annotation in coco_data.get("annotations", [])
            if annotation.get("attributes", {}).get("class") == "PaintDamage"
        ]

        filtered_coco_data = {
            "info": coco_data.get("info", {}),
            "licenses": coco_data.get("licenses", []),
            "images": coco_data.get("images", []),
            "categories": coco_data.get("categories", []),
            "annotations": filtered_annotations,
        }

        with open(output_json_path, 'w') as f:
            json.dump(filtered_coco_data, f, ensure_ascii=False, indent=4)

    except Exception as e:
        print(f"{json_file} 처리 중 오류 발생: {e}")

print("모든 JSON 파일 처리 완료!")
```

```python
import os
from PIL import Image

def convert_tiff_to_jpg(image_dir):
    """
    image_dir에 있는 모든 TIFF 이미지를 JPG로 변환하고,
    변환 후 원본 TIFF 파일을 삭제합니다.

    Args:
      image_dir: 이미지 파일 디렉토리
    """
    for filename in os.listdir(image_dir):
        if filename.endswith(".tiff"):
            tiff_path = os.path.join(image_dir, filename)
            jpg_path = os.path.join(image_dir, filename.replace(".tiff", ".jpg"))

            try:
                # TIFF를 JPG로 변환
                with Image.open(tiff_path) as img:
                    img.convert("RGB").save(jpg_path, "JPEG")
                print(f"{tiff_path} -> {jpg_path} 변환 완료")

                # 원본 TIFF 파일 삭제
                os.remove(tiff_path)
                print(f"{tiff_path} 삭제 완료")

            except Exception as e:
                print(f"변환 또는 삭제 오류: {tiff_path}, {e}")

if __name__ == "__main__":
    image_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/images"  # 이미지 파일 디렉토리
    convert_tiff_to_jpg(image_dir)
```

```python
import os
import json

def convert_bbox_to_yolo_format(bbox, image_width, image_height):
    """
    Bounding box 좌표를 YOLO 형식으로 변환합니다.

    Args:
      bbox: Bounding box 좌표 (x_min, y_min, width, height)
      image_width: 이미지 너비
      image_height: 이미지 높이

    Returns:
      YOLO 형식의 bounding box 문자열 (str)
    """
    x_min, y_min, bbox_width, bbox_height = bbox
    x_center = (x_min + bbox_width / 2) / image_width
    y_center = (y_min + bbox_height / 2) / image_height
    width = bbox_width / image_width
    height = bbox_height / image_height
    class_id = 0  # "PaintDamage" 클래스를 0으로 가정
    return f"{class_id} {x_center} {y_center} {width} {height}"


def convert_json_to_yolo(json_path, txt_path):
    """
    JSON 라벨링 데이터를 YOLO 형식의 텍스트 파일로 변환합니다.

    Args:
      json_path: JSON 라벨링 파일 경로
      txt_path: YOLO 형식의 텍스트 파일로 저장할 경로
    """
    try:
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # 이미지 정보 가져오기
        image_width = data["images"][0]["width"]
        image_height = data["images"][0]["height"]

        with open(txt_path, 'w') as f:
            for annotation in data["annotations"]:
                bbox = annotation["bbox"]
                yolo_label = convert_bbox_to_yolo_format(bbox, image_width, image_height)
                f.write(f"{yolo_label}\n")
        print(f"{json_path} -> {txt_path} 변환 완료")

    except Exception as e:
        print(f"변환 오류: {json_path}, {e}")


if __name__ == "__main__":
    json_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/laveling"  # JSON 파일 디렉토리

    for filename in os.listdir(json_dir):
        if filename.endswith(".json"):
            json_path = os.path.join(json_dir, filename)
            txt_path = os.path.join(json_dir, filename.replace(".json", ".txt"))

            convert_json_to_yolo(json_path, txt_path)

            # 변환 후 JSON 파일 삭제
            try:
                os.remove(json_path)
                print(f"{json_path} 삭제 완료")
            except Exception as e:
                print(f"{json_path} 삭제 오류: {e}")
```

```python
import os
import shutil
import random

def split_dataset(image_dir, label_dir, train_dir, val_dir, train_ratio=0.8):
  """
  데이터셋을 train과 validation으로 분리합니다.

  Args:
    image_dir: 이미지 파일 디렉토리
    label_dir: 라벨 파일 디렉토리
    train_dir: train 데이터셋을 저장할 디렉토리
    val_dir: validation 데이터셋을 저장할 디렉토리
    train_ratio: train 데이터셋 비율 (default: 0.8)
  """

  # 디렉토리 생성
  os.makedirs(train_dir, exist_ok=True)
  os.makedirs(val_dir, exist_ok=True)

  # 이미지 파일 목록 가져오기
  image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg')]
  random.shuffle(image_files)  # 파일 목록 섞기

  # train/val 분리
  train_num = int(len(image_files) * train_ratio)
  train_files = image_files[:train_num]
  val_files = image_files[train_num:]

  # train 데이터셋 이동
  for filename in train_files:
    image_path = os.path.join(image_dir, filename)
    label_filename = filename.replace(".jpg", ".txt")
    label_path = os.path.join(label_dir, label_filename)

    shutil.move(image_path, train_dir)
    shutil.move(label_path, train_dir)

  # validation 데이터셋 이동
  for filename in val_files:
    image_path = os.path.join(image_dir, filename)
    label_filename = filename.replace(".jpg", ".txt")
    label_path = os.path.join(label_dir, label_filename)

    shutil.move(image_path, val_dir)
    shutil.move(label_path, val_dir)

  print(f"Train: {len(train_files)}개, Validation: {len(val_files)}개 분리 완료")


if __name__ == "__main__":
  image_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/images"
  label_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/laveling"
  train_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/train"
  val_dir = "/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/val"
  split_dataset(image_dir, label_dir, train_dir, val_dir)
```

```python
import yaml

def create_yaml(train_dir, val_dir, nc, names):
  """
  YOLOv8 모델 학습에 필요한 data.yaml 파일을 생성합니다.

  Args:
    train_dir: 학습 데이터셋 경로 (str)
    val_dir: 검증 데이터셋 경로 (str)
    nc: 클래스 개수 (int)
    names: 클래스 이름 리스트 (list)
  """
  data_dict = {
      'train': train_dir,
      'val': val_dir,
      'nc': nc,
      'names': names
  }

  # data.yaml 파일 저장 경로
  save_path = '/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/yolo_data.yaml'  # 원하는 경로로 변경

  with open(save_path, 'w') as f:
      yaml.dump(data_dict, f, indent=2)

  print(f"data.yaml 파일이 {save_path}에 생성되었습니다.")

# 사용 예시
train_dir = '/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/train'  # train 폴더 경로
val_dir = '/content/drive/MyDrive/Homeplus/data/Steel/CoatingDamage/datasets/val'  # val 폴더 경로
nc = 1  # 클래스 개수 (PaintDamage 1개)
names = ['PaintDamage']  # 클래스 이름 리스트

create_yaml(train_dir, val_dir, nc, names)
```
