```python
import os
import json
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# 경로 설정
image_folder = '/content/콘크리트_철근노출_원천_01/'  # TIFF 이미지 파일 폴더
json_folder = '/content/콘크리트_철근노출_라벨링_01/'  # JSON 파일 폴더

# Step 1: TIFF 이미지 파일 랜덤으로 5개 선택
image_files = [f for f in os.listdir(image_folder) if f.endswith('.tiff')]
selected_images = random.sample(image_files, min(len(image_files), 5))  # 랜덤으로 최대 5개 선택

# Step 2: 이미지와 JSON 내용을 시각적으로 출력
for image_file in selected_images:
    base_name = os.path.splitext(image_file)[0]  # 확장자를 제외한 파일 이름
    json_file = os.path.join(json_folder, base_name + '.json')
    image_path = os.path.join(image_folder, image_file)

    # TIFF 이미지 출력
    img = Image.open(image_path)
    fig, ax = plt.subplots(1, figsize=(8, 8))
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(f"{image_file}")

    # JSON 파일에서 bounding box 가져오기
    if os.path.exists(json_file):
        with open(json_file, 'r') as f:
            json_data = json.load(f)

        annotations = json_data.get('annotations', [])  # 모든 annotations 사용

        # Bounding box 그리기
        for annotation in annotations:
            bbox = annotation.get('bbox', [])
            attributes = annotation.get('attributes', {})
            class_name = attributes.get('class', 'Unknown')  # attributes 안에 class 가져오기

            if len(bbox) == 4:  # bbox는 [x, y, width, height] 형식이어야 함
                x, y, width, height = bbox
                rect = patches.Rectangle(
                    (x, y), width, height, linewidth=2, edgecolor='red', facecolor='none'
                )
                ax.add_patch(rect)
                ax.text(
                    x, y - 5, class_name,  # class_name 표시
                    color='red', fontsize=10, bbox=dict(facecolor='white', alpha=0.5)
                )
    else:
        print(f"{base_name}.json 파일이 없습니다.")

    plt.show()
```

```python
# 경로 설정
json_folder = '/content/콘크리트_철근노출_라벨링_01/'  # JSON 파일이 있는 폴더
output_folder = '/content/exposure_json/'  # 필터링된 JSON 저장 폴더

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# JSON 파일 처리
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)
        output_json_path = os.path.join(output_folder, json_file)  # 필터링 결과 저장 경로

        # JSON 데이터 로드
        with open(input_json_path, 'r') as f:
            coco_data = json.load(f)

        # 필터링 작업: Exposure 클래스만 포함
        filtered_annotations = [
            annotation for annotation in coco_data.get("annotations", [])
            if annotation.get("attributes", {}).get("class") == "Exposure"
        ]

        # 필터링된 데이터 구성
        filtered_coco_data = {
            "info": coco_data.get("info", {}),
            "licenses": coco_data.get("licenses", []),
            "images": coco_data.get("images", []),  # 모든 이미지 유지
            "categories": coco_data.get("categories", []),
            "annotations": filtered_annotations,  # Exposure 클래스만 포함
        }

        # 필터링된 JSON 저장
        with open(output_json_path, 'w') as f:
            json.dump(filtered_coco_data, f, ensure_ascii=False, indent=4)

        print(f"{json_file} 처리 완료! 필터링된 결과가 {output_json_path}에 저장되었습니다.")

print("모든 JSON 파일 처리 완료!")
```

```python
from collections import defaultdict

# 경로 설정
json_folder = '/content/exposure_json/'  # JSON 파일이 있는 폴더

# 클래스별 개수를 저장할 딕셔너리
class_counts = defaultdict(int)

# JSON 파일 순회
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)

        # JSON 데이터 로드
        with open(input_json_path, 'r') as f:
            coco_data = json.load(f)

        # annotations에서 class 개수 카운트
        for annotation in coco_data.get("annotations", []):
            class_name = annotation.get("attributes", {}).get("class")
            if class_name:  # 클래스가 존재할 경우만 카운트
                class_counts[class_name] += 1

# 결과 출력
print("존재하는 클래스와 각 클래스의 개수:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}개")

# 총 클래스 종류
print(f"\n총 클래스 종류: {len(class_counts)}개")
```

```python
# 경로 설정
json_folder = '/content/콘크리트_철근노출_라벨링_02/'  # JSON 파일이 있는 폴더
output_folder = '/content/exposure_json_02/'  # 필터링된 JSON 저장 폴더

# 출력 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# JSON 파일 처리
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)
        output_json_path = os.path.join(output_folder, json_file)  # 필터링 결과 저장 경로

        try:
            # JSON 데이터 로드
            with open(input_json_path, 'r', encoding='utf-8') as f:
                coco_data = json.load(f)

            # 필터링 작업: Exposure 클래스만 포함
            filtered_annotations = [
                annotation for annotation in coco_data.get("annotations", [])
                if annotation.get("attributes", {}).get("class") == "Exposure"
            ]

            # 필터링된 데이터 구성
            filtered_coco_data = {
                "info": coco_data.get("info", {}),
                "licenses": coco_data.get("licenses", []),
                "images": coco_data.get("images", []),  # 모든 이미지 유지
                "categories": coco_data.get("categories", []),
                "annotations": filtered_annotations,  # Exposure 클래스만 포함
            }

            # 필터링된 JSON 저장
            with open(output_json_path, 'w', encoding='utf-8') as f:
                json.dump(filtered_coco_data, f, ensure_ascii=False, indent=4)

            print(f"{json_file} 처리 완료! 필터링된 결과가 {output_json_path}에 저장되었습니다.")

        except UnicodeDecodeError as e:
            print(f"파일 {json_file} 처리 중 오류 발생: {e}")
        except json.JSONDecodeError as e:
            print(f"파일 {json_file}의 JSON 형식 오류: {e}")
        except Exception as e:
            print(f"파일 {json_file} 처리 중 예상치 못한 오류 발생: {e}")

print("모든 JSON 파일 처리 완료!")
```

```python
from collections import defaultdict
import os
import json

# 경로 설정
json_folder = '/content/exposure_json_02/'  # JSON 파일이 있는 폴더

# 클래스별 개수를 저장할 딕셔너리
class_counts = defaultdict(int)

# JSON 파일 순회
for json_file in os.listdir(json_folder):
    if json_file.endswith('.json'):  # JSON 파일만 처리
        input_json_path = os.path.join(json_folder, json_file)

        try:
            # JSON 데이터 로드
            with open(input_json_path, 'r', encoding='utf-8') as f:
                coco_data = json.load(f)

            # annotations에서 class 개수 카운트
            for annotation in coco_data.get("annotations", []):
                class_name = annotation.get("attributes", {}).get("class")
                if class_name:  # 클래스가 존재할 경우만 카운트
                    class_counts[class_name] += 1

        except UnicodeDecodeError as e:
            print(f"파일 {json_file} 처리 중 UnicodeDecodeError 발생: {e}")
        except json.JSONDecodeError as e:
            print(f"파일 {json_file}의 JSON 형식 오류: {e}")
        except Exception as e:
            print(f"파일 {json_file} 처리 중 예상치 못한 오류 발생: {e}")

# 결과 출력
print("존재하는 클래스와 각 클래스의 개수:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}개")

# 총 클래스 종류
print(f"\n총 클래스 종류: {len(class_counts)}개")
```

## 데이터셋 불러오기

```python
import os
import json
from PIL import Image

# 라벨링 데이터 및 원천 데이터 디렉토리 경로 리스트
label_dirs = [
    "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/라벨링데이터/exposure_json_01",
    "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/라벨링데이터/exposure_json_02"
]
source_dirs = [
    "/content/콘크리트_철근노출_원천_01",
    "/content/콘크리트_철근노출_원천_02"
]

# 라벨링 데이터 불러오기
def load_label_data(label_dirs):
    label_data = {}
    for label_dir in label_dirs:
        for file_name in os.listdir(label_dir):
            if file_name.endswith('.json'):
                file_path = os.path.join(label_dir, file_name)
                try:
                    with open(file_path, 'r', encoding='utf-8') as json_file:
                        label_data[file_name] = json.load(json_file)
                except Exception as e:
                    print(f"라벨링 데이터 로드 실패: {file_name}, 오류: {e}")
    return label_data

# 원천 데이터 불러오기
def load_source_data(source_dirs):
    source_data = []
    for source_dir in source_dirs:
        for file_name in os.listdir(source_dir):
            if file_name.endswith(('.tiff')):
                file_path = os.path.join(source_dir, file_name)
                try:
                    image = Image.open(file_path)
                    source_data.append((file_name, image))
                except Exception as e:
                    print(f"이미지 로드 실패: {file_name}, 오류: {e}")
    return source_data

# 데이터 로드
label_data = load_label_data(label_dirs)
source_data = load_source_data(source_dirs)

# 결과 출력
print(f"라벨링 데이터 수: {len(label_data)}")
print(f"원천 데이터 수: {len(source_data)}")
```

```python
# 라벨링 데이터와 원천 데이터 이름 일치 확인
def check_label_source_matching(label_dirs, source_dirs):
    missing_files = []  # 라벨링에서 참조하지만 원천 데이터에 없는 파일 목록
    matched_files = []  # 라벨링과 원천 데이터가 일치하는 파일 목록

    # 원천 데이터 파일 이름 리스트 생성
    source_file_names = set()
    for source_dir in source_dirs:
        for file_name in os.listdir(source_dir):
            if file_name.endswith(('.tiff')):
                source_file_names.add(file_name)

    # 라벨링 데이터에서 참조하는 이미지 이름 확인
    for label_dir in label_dirs:
        for file_name in os.listdir(label_dir):
            if file_name.endswith('.json'):
                file_path = os.path.join(label_dir, file_name)
                with open(file_path, 'r', encoding='utf-8') as f:
                    json_data = json.load(f)

                # JSON 파일에서 참조하는 이미지 이름
                image_name = json_data["images"][0]["file_name"]

                # 원천 데이터에 해당 이미지가 있는지 확인
                if image_name in source_file_names:
                    matched_files.append(image_name)
                else:
                    missing_files.append(image_name)

    # 결과 출력
    print(f"일치하는 파일 수: {len(matched_files)}")
    print(f"누락된 파일 수: {len(missing_files)}")

    if missing_files:
        print("\n누락된 파일 목록:")
        for missing_file in missing_files:
            print(missing_file)

# 함수 실행
check_label_source_matching(label_dirs, source_dirs)
```

```python
import os
import json
import random
from PIL import Image

label_dirs = [
    "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/라벨링데이터/exposure_json_01",
    "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/라벨링데이터/exposure_json_02"
]
source_dirs = [
    "/content/콘크리트_철근노출_원천_01",
    "/content/콘크리트_철근노출_원천_02"
]


# 데이터셋 저장 경로
dataset_dir = "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/dataset"

# 데이터셋 저장 디렉토리 설정
train_images_dir = os.path.join(dataset_dir, "images", "train")
val_images_dir = os.path.join(dataset_dir, "images", "val")
train_labels_dir = os.path.join(dataset_dir, "labels", "train")
val_labels_dir = os.path.join(dataset_dir, "labels", "val")

# 디렉토리 생성
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# YOLO 라벨 변환 함수
def convert_bbox_to_yolo_format(bbox, image_width, image_height):
    x_min, y_min, bbox_width, bbox_height = bbox
    x_center = (x_min + bbox_width / 2) / image_width
    y_center = (y_min + bbox_height / 2) / image_height
    width = bbox_width / image_width
    height = bbox_height / image_height
    return f"0 {x_center} {y_center} {width} {height}"  # class_id는 0으로 설정

# 데이터셋 구축 함수
def create_yolo_dataset(label_dirs, source_dirs, train_ratio=0.8):
    json_files = []

    # 모든 JSON 파일과 원본 이미지 파일 매칭
    for label_dir, source_dir in zip(label_dirs, source_dirs):
        for file_name in os.listdir(label_dir):
            if file_name.endswith(".json"):
                json_files.append((label_dir, source_dir, file_name))

    random.shuffle(json_files)  # 데이터 섞기

    train_count = int(len(json_files) * train_ratio)
    train_files = json_files[:train_count]
    val_files = json_files[train_count:]

    train_image_count, val_image_count = 0, 0
    train_label_count, val_label_count = 0, 0

    for label_dir, source_dir, json_file in json_files:
        json_path = os.path.join(label_dir, json_file)
        try:
            with open(json_path, 'r', encoding='utf-8') as f:
                data = json.load(f)

            # 이미지 정보 가져오기
            image_info = data["images"][0]
            image_path = os.path.join(source_dir, image_info["file_name"])
            image_width, image_height = image_info["width"], image_info["height"]

            # 저장 경로 설정
            if (label_dir, source_dir, json_file) in train_files:
                image_save_dir = train_images_dir
                label_save_dir = train_labels_dir
                train_image_count += 1
            else:
                image_save_dir = val_images_dir
                label_save_dir = val_labels_dir
                val_image_count += 1

            # 이미지 저장 (.tiff -> .jpg 확장자 변경)
            dest_image_path = os.path.join(image_save_dir, image_info["file_name"].replace(".tiff", ".jpg"))
            if not os.path.exists(dest_image_path):
                with Image.open(image_path) as img:
                    img.convert("RGB").save(dest_image_path, "JPEG")

            # YOLO 라벨 저장
            label_save_path = os.path.join(label_save_dir, image_info["file_name"].replace(".tiff", ".txt"))
            with open(label_save_path, 'w') as f:
                for annotation in data["annotations"]:
                    bbox = annotation["bbox"]
                    yolo_label = convert_bbox_to_yolo_format(bbox, image_width, image_height)
                    f.write(f"{yolo_label}\n")
                    if (label_dir, source_dir, json_file) in train_files:
                        train_label_count += 1
                    else:
                        val_label_count += 1

        except Exception as e:
            print(f"오류 발생: {json_file}, {e}")

    # 결과 출력
    print(f"Train 데이터셋 - 이미지: {train_image_count}, 라벨: {train_label_count}")
    print(f"Validation 데이터셋 - 이미지: {val_image_count}, 라벨: {val_label_count}")

# 데이터셋 생성 실행
create_yolo_dataset(label_dirs, source_dirs)
```

```python
from ultralytics import YOLO

# YOLO 모델 초기화
model = YOLO('yolo11m.pt')  # 사전 학습된 YOLOv11n 모델

# 모델 학습 실행
model.train(
    data='/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/dataset/dataset.yaml',  # yaml 파일 경로
    epochs=50,                     # 학습할 에포크 수
    batch=16,                      # 배치 크기
    imgsz=640,                     # 이미지 크기
    project='/content/yolo_training',  # 학습 결과 저장 경로
    name='concrete_spalling_v1',       # 실험 이름
    workers=2,                     # 데이터 로드 워커 수
    optimizer='AdamW',             # 옵티마이저 변경 (AdamW 추천)
    lr0=0.001,                     # 초기 학습률
    cos_lr=True,                   # 코사인 학습률 스케줄러 사용
    augment=True,                  # 데이터 증강 활성화
    mosaic=1.0                     # 모자이크 증강
)
```

```python
import shutil
import os

# YOLO 학습 결과 경로
source_path = "/content/yolo_training/concrete_spalling_v1"

# Google Drive 저장 경로
destination_path = "/content/drive/MyDrive/Homeplus/data/Concrete/ExposedReinforcement/concrete_spalling_v1"

# 대상 폴더 생성
os.makedirs(destination_path, exist_ok=True)

# YOLO 결과 파일 복사
for item in os.listdir(source_path):
    src_item = os.path.join(source_path, item)
    dest_item = os.path.join(destination_path, item)

    # 파일이나 폴더를 구분하여 복사
    if os.path.isfile(src_item):
        shutil.copy2(src_item, dest_item)
    elif os.path.isdir(src_item):
        shutil.copytree(src_item, dest_item, dirs_exist_ok=True)

print(f"YOLO 학습 결과가 {destination_path}에 저장되었습니다.")
```
